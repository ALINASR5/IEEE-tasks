# Clean Data

```
# Shape before cleaning
print(df.shape)
```

```
# Shape after drop_duplicates
df = df.drop_duplicates()
print(df.shape)
```

```
# check nulls
print("\nNull values:\n", df.isnull().sum())
```

```
# check types
print("\nDtypes:\n", df.dtypes)
```

```
# Cast Dates
date_cols = ['session_date', 'signup_date', 'last_login_date']
df[date_cols] = df[date_cols].apply(pd.to_datetime)
```

```
# check neumeric range
print("\nage min:", df['age'].min(), "| max:", df['age'].max())
print("session_duration_minutes min:", df['session_duration_minutes'].min(), "| max:", df['session_duration_minutes'].max())
print("calories_burned min:", df['calories_burned'].min(), "| max:", df['calories_burned'].max())
```

```
#check categoricals
cat_cols = ['gender', 'workout_module_used', 'completed_goal_checkin', 'dropoff_flag', 'goal']
for col in cat_cols:
    print(f"\n{col} unique values: {df[col].unique()}")
```

```
df['completed_goal_checkin'] = df['completed_goal_checkin'].map({'Yes': 1, 'No': 0})
df['dropoff_flag'] = df['dropoff_flag'].map({'Yes': 1, 'No': 0})
df['gender'] = df['gender'].map({'M': 1, 'F': 0})
```

```
df = pd.get_dummies(df, columns=['workout_module_used', 'goal'])
```

```
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```
df.columns = df.columns.str.replace(' ', '_')
```

```
# Final check
print("\nshape:", df.shape)
print("\ntypes:\n", df.dtypes)
print("\nhead:\n", df.head())
```
